## Looking at Skewness and Kurtosis

```python
training_data.skew(), training_data.kurt()
```

```python
y = training_data['SalePrice']
plt.figure(1); plt.title('Johnson SU')
sns.distplot(y, kde=False, fit=st.johnsonsu)
plt.figure(2); plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)
plt.figure(3); plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
```

From the graphs, it is evident that this is not a normal distribution. We will have to transform the data before performing a regression. The best fit is unbounded Johnson distribution.

```python
sns.distplot(training_data.skew(),color='blue',axlabel='Skewness')
```

```python
plt.figure(figsize = (12,8))
sns.distplot(training_data.kurt(),color='r',axlabel ='Kurtosis',norm_hist= False, kde = True,rug = False)
plt.show()
```

## Zoomed HeatMap

```python
k = 11
cols = correlation.nlargest(k,'SalePrice')['SalePrice'].index
print(cols)
cm = np.corrcoef(training_data[cols].values.T)
f, ax = plt.subplots(figsize = (14, 12))
sns.heatmap(cm, vmax=.8, linewidths=0.01, square=True, annot=True, cmap='viridis',
           linecolor='white',xticklabels = cols.values, annot_kws = {'size':12},yticklabels = cols.values)
```
